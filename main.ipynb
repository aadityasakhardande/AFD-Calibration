```python
import pandas as pd
import matplotlib.pyplot as plt
import glob
import ipywidgets as widgets
import matplotlib.pyplot as plt                 
import ipywidgets as widgets
import numpy as np

from IPython.display import display

# Set a clean plot style
plt.style.use('ggplot')
```

```python
# Find and load the first 10 CSVs
all_files = sorted(glob.glob("BWP_Data_CSV/*.csv"))[:25]
data_dict = {}

for f in all_files:
    data_dict[f] = pd.read_csv(
        f, 
        header=None, 
        names=["time", "Position", "Force", "Error", "Commanded Force", "Tool Status"]
    )

print(f"Loaded {len(data_dict)} files.")
```

```python


def overlay_commanded_force(selected_files):
    plt.figure(figsize=(14,6))
    
    for filename in selected_files:
        df = data_dict[filename]
        
        # X-axis = point index
        x = np.arange(len(df))
        y = df["Commanded Force"]
        
        plt.plot(x, y, label=filename)
    
    plt.title("Commanded Force vs Sample Index")
    plt.xlabel("Sample Number")
    plt.ylabel("Commanded Force")
    plt.grid(True)
    plt.legend()
    plt.show()


widgets.interact(
    overlay_commanded_force,
    selected_files=widgets.SelectMultiple(
        options=list(data_dict.keys()),
        description='Select Runs',
        rows=10
    )
);
```

```python
def overlay_force(selected_files):
    plt.figure(figsize=(14,6))
    
    for filename in selected_files:
        df = data_dict[filename]
        
        # X-axis = point index
        x = np.arange(len(df))
        y = df["Force"]
        
        plt.plot(x, y, label=filename)
    
    plt.title("Force vs Sample Index")
    plt.xlabel("Sample Number")
    plt.ylabel("Force")
    plt.grid(True)
    plt.legend()
    plt.show()


widgets.interact(
    overlay_force,
    selected_files=widgets.SelectMultiple(
        options=list(data_dict.keys()),
        description='Select Runs',
        rows=10
    )
);
```

```python
def overlay_position(selected_files):
    plt.figure(figsize=(14,6))
    
    for filename in selected_files:
        df = data_dict[filename]
        
        # X-axis = point index
        x = np.arange(len(df))
        y = df["Position"]
        
        plt.plot(x, y, label=filename)
    
    plt.title("Position vs Sample Index")
    plt.xlabel("Sample Number")
    plt.ylabel("Position")
    plt.grid(True)
    plt.legend()
    plt.show()


widgets.interact(
    overlay_position,
    selected_files=widgets.SelectMultiple(
        options=list(data_dict.keys()),
        description='Select Runs',
        rows=10
    )
);
```

```python

def plot_force_vs_cmd(selected_files):
    for filename in selected_files:
        df = data_dict[filename]
        x = np.arange(len(df))
        
        plt.figure(figsize=(12,5))
        plt.plot(x, df["Force"], label="Force")
        plt.plot(x, df["Commanded Force"], label="Commanded Force")
        
        plt.title(f"{filename} - Force & Commanded Force vs Sample Index")
        plt.xlabel("Sample Number")
        plt.ylabel("Force")
        plt.grid(True)
        plt.legend()
        plt.show()


widgets.interact(
    plot_force_vs_cmd,
    selected_files=widgets.SelectMultiple(
        options=list(data_dict.keys()),
        description="Select CSVs",
        rows=10
    )
);
```

```python
def plot_error(selected_files):
    for filename in selected_files:
        df = data_dict[filename]
        x = np.arange(len(df))
        
        plt.figure(figsize=(12,5))
        plt.plot(x, df["Error"], label="Error", color="red")
        
        plt.title(f"{filename} - Error vs Sample Index")
        plt.xlabel("Sample Number")
        plt.ylabel("Error")
        plt.grid(True)
        plt.legend()
        plt.show()


widgets.interact(
    plot_error,
    selected_files=widgets.SelectMultiple(
        options=list(data_dict.keys()),
        description="Select CSVs",
        rows=10
    )
);
```
